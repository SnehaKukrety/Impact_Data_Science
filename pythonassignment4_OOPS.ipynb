```python
#Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
'''A class is a blueprint or a template for creating objects. It defines the properties and behaviors that objects of
the class will possess. It acts as a container for data (attributes) and functions (methods) that operate on that data. 
In simpler terms, a class describes the structure and behavior of objects.It is defined using the keyword class.
Syntax for creating a class:
class class_name:

An object, on the other hand, is an instance of a  class. It is a specific entity created based on the class definition.
Objects have their own unique identity and can store data specific to their instance, as well as perform actions defined 
by the class methods.Syntax for creating object:
object_name=class_name()

Example:'''
class A:
    def __init__(self,name):
        self.name=name
    def display(self):    
        print(f'{self.name} Welcome to class A')
obj_1=A("Sneha Kukrety")
obj_2=A("Raman Lamba")
obj_1.display()
obj_2.display()

        
```

```python
#Q2. Name the four pillars of OOPs.
'''There are four major pillars in oops.Inheritance, Polymorphism, Encapsulation and Abstraction.
1) Inheritance: Itis a fundamental concept in object-oriented programming (OOP) that allows classes to inherit
attributes and methods from other classes.It promotes code reuse, extensibility, and hierarchical organization of classes.
There are 5 types of inheritance: Single inheritance, multiple inheritance, multilevel inheritance, hierarchical
inheritance, hybrid inheritance. Example:'''

class A:
    def hello(self):
        print("Welcome to class A")
class B(A):
    def display(self):
        print("Welcome to class B")
class C:
    def Welcome(self):
        print("Welcome to class C")
class D(A,C):
    def display():
        pass

obj_b=B()
obj_d=D()
print(obj_b.hello())
print(obj_b.display())
print(obj_d.Welcome())
print(obj_d.hello())

'''2) Polymorphism means many forms. In this overloading and over-riding is performed. Overloading is a compile time 
polymorphism in which functions have same name(in parents and child class or even within the same class) but different 
arguments are given.In over-riding the function name and arguments are of same type (in parent and child class). It is a run
time polymorphism. Example:'''

class Shape:
    def calculate_area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

circle = Circle(5)
rectangle = Rectangle(4, 6)

print(circle.calculate_area())   

print(rectangle.calculate_area())
 
    

'''3)Encapsulation and 4) Abstraction:
Encapsulation is the practice of bundling data and the methods that operate on that data within a single unit called a class.
Abstraction is the process in which hiding the implementation and showing only the features'''  
class BankAccount:
    def __init__(self, account_number, balance):
        self._acc_number = account_number
        self._balance = balance

    def deposit(self, amount):
        self._balance += amount

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
        else:
            print("Insufficient balance.")

    def get_balance(self):
        return self._balance


class Customer:
    def __init__(self, name, account):
        self._name = name
        self._account = account

    def display_balance(self):
        print(f"Account balance: {self._account.get_balance()}")


account = BankAccount("1234567890", 1000)
customer = Customer("Sneha Kukrety", account)

customer.display_balance()  

customer._account.withdraw(500) 
customer.display_balance()  

customer._account._balance = 10000  
customer.display_balance()  



```

```python
#Q3. Explain why the __init__() function is used. Give a suitable example.

'''The __init__() function is a constructor in python that is automatically called when an object of the 
class is created. It is used to initialize the attributes (state) of an object by accepting arguments and
assigning them to the object's attributes.Example:'''
class A:
    def __init__(self,name):
        self.name=name
    def display(self):    
        print(f'{self.name} Welcome to class A')
obj_1=A("Sneha Kukrety")
obj_2=A("Raman Lamba")
obj_1.display()
obj_2.display()
```

```python
#Q4. Why self is used in OOPs?
'''The self parameter is used to refer to the current instance of a class. It is a convention in Python 
to name the first parameter of instance methods as self. The purpose of self is to allow
access to the attributes and methods of the object within its own class.'''
class A:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, Welcome {self.name}. Nice to meet you!!!")


# Creating two instances of the Person class
person1 = A("Sneha")
person2 = A("Priyanshu")

person1.greet()  
person2.greet()  
```

```python
#Q5. What is inheritance? Give an example for each type of inheritance.
'''Inheritance is a fundamental concept in object-oriented programming (OOP) that allows classes to inherit 
attributes and methods from other classes. It promotes code reuse, extensibility, and hierarchical organization of classes.
There are 5 types of inheritance: Single inheritance, multiple inheritance, multilevel inheritance, hierarchical
inheritance, hybrid inheritance. Example:'''

#single inheritance

class A:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, Welcome {self.name}. Nice to meet you!!!")

    
class B(A):
    pass
obj=B("Sneha")
print("............Single Inheritance......")
obj.greet()


#Multiple inheritance
class Herbivores:
    def plant(self):
        print("Eat plants")

class Carnivores:
    def Meat(self):
        print("Eat meat")

class Omnivores(Herbivores, Carnivores):
    def both(self):
        print("Eat both")
        
print("..............Multiple Inheritance......")
omni=Omnivores()
omni.plant() 
omni.Meat  
omni.both()  

#Multilevel Inheritance
class University:
    def Dept(self):
        print("Department")

class College(University):
    def Display(self):
        print("Welcome to college")

class Class(College):
    def Student(self):
        print("Welcome to class")

print(".............Multilevel Inheritance......")
obj=Class()
obj.Dept()
obj.Display()
obj.Student()


#hierchical Inheritance    
class Animal:
    def eat(self):
        print("Animal is eating.")

class Dog(Animal):
    def bark(self):
        print("Dog is barking.")

class Cat(Animal):
    def meow(self):
        print("Cat is meowing.")
print("..............Hierchical Inheritance.........")
dog = Dog()
dog.eat()  
dog.bark() 

cat = Cat()
cat.eat() 
cat.meow()  

#Hybrid Inheritance
class Car:
    def display(self):
        print("It is a car")


class Petrol(Car):
    def petrol_car(self):
        print("Petrol Car")


class Diesel(Car):
    def diesel_car(self):
        print("Diesel car")


class Bike(Petrol, Diesel):
    def navigate(self):
        print("Bike")

print("................Hybrid Inheritance........")
bike = Bike()
bike.display()
bike.petrol_car() 
bike.diesel_car() 
bike.navigate() 

```
